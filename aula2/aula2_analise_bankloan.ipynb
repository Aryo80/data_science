*Categorias*

Vamos indicar para o Pandas que a coluna educação tem dados qualitativos

```python
dados.EDUCACAO = dados.EDUCACAO.astype('category')
```

```python
dados.EDUCACAO
```

Vamos substituir os números das categorias por rótulos

```python
dados.EDUCACAO.cat.categories = ('Analfabeto', '1o. Grau', '2o. Grau', 'Graduacao', 'Pos-Graduacao')
```

```python
dados
```

```python
dados.DEFAULT = dados.DEFAULT.astype('category')
dados.DEFAULT.cat.categories = ('Adimplente', 'Inadimplente')
```

```python
dados
```

==Gráficos==

O pandas usa os recursos de gráficos do _matplotlib_

As linhas abaixo fazem plots do _matplotlib_ ou pandas aparecerem corretamente no _IPython Notebook_.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
```

Gráfico de barras usando a função [_DataFrame.plot()_](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.plot.html)

```python
pd.DataFrame.plot(kind='bar', data=dados.DEFAULT.value_counts(), color=('blue', 'red'), legend=False)
```

Plot do percentual de adimplentes e inadimplentes.

```python
pd.DataFrame.plot(kind='pie', data=dados.DEFAULT.value_counts(), colors=('blue', 'red'), autopct="%0.2f",legend=False)
```

Gráfico de barras comparando as populações em diversos níveis de educação

```python
pd.DataFrame.plot(kind='bar', data=dados.EDUCACAO.value_counts(), color=('yellow', 'salmon', 'blue', 'red', 'green'),  legend=False)
```

```python
pd.DataFrame.plot(kind='pie', data=dados.EDUCACAO.value_counts(), colors=('yellow', 'salmon', 'blue', 'red', 'green'), autopct="%.2f", legend=False)
```

```python
pd.DataFrame.plot(kind='bar', data=pd.crosstab(dados.EDUCACAO, dados.DEFAULT), color=('yellow', 'salmon'), stacked=True)
```

```python
pd.DataFrame.plot(kind='bar', data=pd.crosstab(dados.DEFAULT, dados.EDUCACAO), color=('yellow', 'salmon', 'blue', 'red', 'green'), stacked=True)
```

```python
pd.DataFrame.plot(kind='bar', data=pd.crosstab(dados.DEFAULT, dados.EDUCACAO), color=('yellow', 'salmon', 'blue', 'red', 'green'), legend=False, stacked=True)
```

```python
pd.DataFrame.plot(kind='pie', data=dados.EDUCACAO[dados.DEFAULT == 'Adimplente'].value_counts(), autopct="%.2f",legend=False)
```

```python
pd.DataFrame.plot(kind='pie', data=dados.EDUCACAO[dados.DEFAULT == 'Inadimplente'].value_counts(), legend=False)
```
